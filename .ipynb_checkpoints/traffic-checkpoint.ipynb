```python
import pandas as pd
%matplotlib inline
df=pd.read_csv("traffic.csv")##   ye ol' imports
```

```python
df.drop(labels=['id','device_id','record_oid','count_start_date','count_end_date','max_speed','speed95_percent','speed85_percent','counter_type','counter_number','council_district','average_daily_bike_traffic','ward','tract','public_works_division','pli_division','police_zone','fire_zone'], axis=1,inplace= True)
df.dropna(subset=['average_daily_car_traffic'], inplace=True)##purging columns and NaN values
col = df.pop('neighborhood')
df.set_index(col, inplace=True)##moving neighborhood column to index
df.head()
```

```python
df.index.name## was having some trouble with my index name so I checked
pdf = df.drop(labels=['longitude','latitude'], axis=1)##didn't want average location
mdf = pdf.groupby(level='neighborhood').mean()
mdf.sort_values(['average_daily_car_traffic'], inplace=True)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.precision',1):
    print(mdf[['average_daily_car_traffic','median_speed']])##selected these two just for ease of the eyes (I'll get into why I picked median soon)
```
